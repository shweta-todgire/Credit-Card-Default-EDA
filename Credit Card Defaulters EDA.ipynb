**Default of Credit Card Clients Dataset**

The Default of Credit Card Clients Dataset contains information about 30,000 credit card customers in Taiwan. It includes customer demographics, credit limits, payment history, bill amounts, payment amounts, and default status. This dataset is widely used for credit risk analysis and contains 30,000 rows and 25 columns.

```python
# Import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load & Inspect the data

df = pd.read_csv("Default of Credit Cards Clients.csv")

print("First 5 Rows")
display(df.head())

print("Dataset Shape:")
print(df.shape)

print("\nData Types:")
print(df.dtypes)

print("\nSummary Statistics:")
display(df.describe())

print("\nMissing Values Before Cleaning:")
print(df.isnull().sum())

print("\nDuplicate Rows:", df.duplicated().sum())
```

The dataset was checked for missing values, duplicate records, and data types. No missing values or duplicate records were found, so no additional data cleaning needed.

```python
# Descriptive Statistics

numeric_cols = df.select_dtypes(include=np.number)

statistics = pd.DataFrame({
    "Mean": numeric_cols.mean(),
    "Median": numeric_cols.median(),
    "Mode": numeric_cols.mode().iloc[0],
    "Standard Deviation": numeric_cols.std(),
    "Minimum": numeric_cols.min(),
    "Maximum": numeric_cols.max()
})

display(statistics)
```

**Histogram**

```python
plt.figure(figsize=(8,5))
plt.hist(df['AGE'], bins=20, edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

The histogram shows that most credit card customers are between 25 and 40 years old, with the highest age around 30 years. The number of customers decreases as age increases, indicating that younger adults are the majority of the customers, while older customers are fewer.

**Box Plot**

```python
plt.figure(figsize=(6,6))
sns.boxplot(y=df['LIMIT_BAL'], showfliers=False)
plt.title("Credit Limit Distribution")
plt.xlabel("LIMIT_BAL")
plt.ylabel("Credit Limit")

plt.show()
```

The box plot shows that most customers have credit limits between approximately 50,000 and 2,30,000, with the median credit limit around 1,40,000. The distribution is right-skewed, which means a small number of customers have much higher credit limits than the majority. The many points above represent outliers, indicating customers with exceptionally high credit limits (up to 8,00,000). This suggests that while most customers receive moderate credit limits, a few premium customers are granted significantly higher limits.

**Scatter Plot**

```python
plt.figure(figsize=(8,5))
plt.scatter(df['AGE'], df['LIMIT_BAL'], alpha=0.5)
plt.title("Age vs Credit Limit")
plt.xlabel("Age")
plt.ylabel("Credit Limit")
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

This scatter plot shows that there is no strong relationship between age and credit limit. Customers of almost every age have both low and high credit limits. Most customers are between 23 and 60 years old, with credit limits below 5,00,000. A few customers have very high credit limits (up to 1,000,000). This indicates that age alone does not determine a customer's credit limit, and other factors such as income, payment history, or credit score are likely to have a greater influence.

**Bar Chart**

```python
avg_limit = df.groupby('EDUCATION')['LIMIT_BAL'].mean()

plt.figure(figsize=(8,5))
avg_limit.plot(kind='bar')

plt.title("Average Credit Limit by Education")
plt.xlabel("Education")
plt.ylabel("Average Credit Limit")

plt.show()
```

This Bar chart shows that Customers in Education Level 4 receive the highest average credit limit (up to 2,20,000), while Level 3 receives the lowest (up to 1,25,000). Credit limits vary considerably across education groups, which shows that education alone is not a strong predictor of credit allocation and should be analyzed alongside other customer attributes.

**Count Plot**

```python
plt.figure(figsize=(6,4))
sns.countplot(x='default.payment.next.month', data=df)

plt.title("Loan Default Distribution")
plt.xlabel("Default Next Month")
plt.ylabel("Count")

plt.show()
```

Most customers(approx 78%) successfully repaid their loans, while only a small percentage defaulted the next month. This indicates a relatively low default rate.

**Correlation Heatmap**

```python
plt.figure(figsize=(15,10))

sns.heatmap(
    numeric_cols.corr(),
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)

plt.title("Correlation Heatmap")
plt.show()
```

The correlation heatmap indicates that:
- BILL_AMT1–BILL_AMT6 are strongly correlated, showing that customers tend to have similar bill amounts across months.
- PAY_0–PAY_6 are strongly correlated, indicating that customers with payment delays often continue to delay payments.
- PAY_AMT1–PAY_AMT6 show a moderate correlation, suggesting consistent payment behavior over time.
- default.payment.next.month is most closely related to PAY_0–PAY_6, meaning payment delays increase the chance of default.
- AGE, SEX, EDUCATION, and MARRIAGE have a weak correlation with default.payment.next.month, indicating they have little impact on loan default.


```python
# Outliers

Q1 = df['LIMIT_BAL'].quantile(0.25)
Q3 = df['LIMIT_BAL'].quantile(0.75)

IQR = Q3 - Q1

outliers = df[
    (df['LIMIT_BAL'] < Q1 - 1.5 * IQR) |
    (df['LIMIT_BAL'] > Q3 + 1.5 * IQR)
]

print("Number of Outliers:", len(outliers))
display(outliers.head())
```

**Conclusion**

This project shows that customers with payment delays (PAY_0–PAY_6) are more likely to default on their next month's payment, while factors like age, gender, education, and marital status have very little impact. The dataset also contains only 167 outliers out of 30,000 records (about 0.6%), indicating that the data is generally reliable. These insights can help banks identify high-risk customers and make better credit approval and risk management decisions.
